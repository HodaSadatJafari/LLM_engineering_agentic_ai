### PDF Summary

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
aval_api_key=os.getenv("AVALAI_API_KEY")
```

```python
client = OpenAI(
    api_key=aval_api_key,
    base_url="https://api.avalai.ir/v1"
)
```

```python
reader = PdfReader("inputs/HodaSadatJafari_LinkedIn_Profile.pdf")
linkedin = ""
for page in reader.pages:
    text = page.extract_text()
    if text:
        linkedin += text
```

```python
display(Markdown(linkedin))
```

```python
response = client.responses.create(
    model="gpt-4o-mini", 
    input=f"Summerize the following content from LinkedIn profile: {linkedin}",
)

print(response.output_text)
```

### PDF Question Extraction

```python
response = client.responses.create(
    model="gpt-4o-mini", 
    input=f"""Extract 4 questions from the following content.
      
      One Yes/No question.
      Two Four-choice questions.
      One Clarification question.

      Content:
      {linkedin}
    """,
)

print(response.output_text)
```
