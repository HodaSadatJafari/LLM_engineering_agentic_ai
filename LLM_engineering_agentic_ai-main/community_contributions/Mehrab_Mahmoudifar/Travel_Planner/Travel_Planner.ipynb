```python
# Cell 1: Environment Setup
import os
from dotenv import load_dotenv
load_dotenv()

# Cell 2: Search Tool Test
from langchain_community.tools.tavily_search import TavilySearchResults
search = TavilySearchResults(k=2)
print(search.invoke({"query": "weather in Isfahan today"}))

# Cell 3: Graph Execution Test
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

llm = ChatOpenAI(api_key=os.getenv("AVALAI_API_KEY"), base_url="https://api.avalai.ir/v1", model="gpt-4o-mini")
# Simple test run to verify LLM connection
response = llm.invoke([HumanMessage(content="Explain travel benefits in Persian.")])
print(response.content)
```
