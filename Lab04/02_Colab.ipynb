# What is Colab?

Google Colab (short for Colaboratory) is a free cloud-based platform for running Python code, especially popular in data science, machine learning, and AI.

It’s like having a Python notebook in the cloud — no installation required. You can run code directly in your browser.

🔹 Key Features
- Free GPU / TPU access
- You can run heavy machine learning models without a personal GPU.
- Useful for training neural networks, running LLMs, or deep learning experiments.

🔹 Jupyter Notebook Interface
- Colab is essentially a hosted Jupyter notebook.
- You write code in cells and execute them one by one.

🔹 Cloud Storage Integration
- Connect directly to Google Drive to save datasets, models, or notebooks.

🔹 Shareable

- Share notebooks like Google Docs. Others can view or edit.

🔹 Pre-installed libraries
- Comes with Python and many ML/AI libraries like TensorFlow, PyTorch, scikit-learn, pandas, numpy.
- You can also pip install any additional package.
